```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display

df1 = '../data/raw/train/aac/id00017/01dfn2spqyE/00001.npy'

sp1 = np.load(df1)

librosa.display.specshow(librosa.amplitude_to_db(sp1, ref=np.max), y_axis='log', x_axis='time')
plt.title('Power spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
```
